```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# simple pie chart
y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# Add labels to the pie chart with the labels parameter.
y = np.array([35,25,25,15])
mylabels = ["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels = mylabels)
plt.show()
```

```python
# Start Angle
y= np.array([25,35,44,55])
mylabels=["apple","bananas","cherries","dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
# Explode
y = np.array([35,25,25,15])
mylabels = ["Apples","Bananas","Cherries","Dates"]
myexplode = [0.025,0,0,0]
plt.pie(y,labels = mylabels, explode = myexplode)
plt.show()
```

```python
# shadow
y = np.array([35,25,25,15])
mylabels = ["apple","banana","cherries","dates"]
myexplode = [0.1,0,0,0]
plt.pie(y,labels = mylabels, explode = myexplode, shadow =True)
plt.show()
```

```python
# Colors
y = np.array([35,25,25,15])
mylabels = ["Apples","Bananas","Cherries","Dates"]
mycolors = ["black","hotpink","b","#4CAF50"]
plt.pie(y, labels =mylabels, colors = mycolors)
plt.show()
```

```python
# legend
y = np.array([35,25,25,15])
mylabels = ["Apple","Bananas","Cherries","Dates"]
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
# legend with header
y = np.array([35,25,25,15])
mylabels = ["Apples","Bananas","Cherries","Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title ="Four Fruits:",loc="upper right",fontsize='small',bbox_to_anchor=(1.25, 1.25))
plt.show()
```
